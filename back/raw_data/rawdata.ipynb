```python
import pandas as pd
dfDeposit = pd.read_csv("deposit_data_csv.csv")
dfDeposit.head() # 데이터프레임(df) dfDeposit의 상위 5개의 행을 출력
```

```python
dfDeposit.shape # df의 행(row)과 열(column) 수를 튜플 형태로 반환
# => 39행 15열
```

```python
# isnull() : df에서 결측값(Null 값)이 있는지 확인 -> 결측값인 위치를 True로 표시한 동일 크기의 df를 반환
# sum() : 각 열(column)별로 True 값(결측값)의 개수를 합산
dfDeposit.isnull().sum()
```

```python
'''
통계량 /	의미

count        : 각 열에서 값이 존재하는 데이터 개수 (결측값 제외)
mean         : 평균값
std	         : 표준편차(Standard Deviation)
min	         : 최솟값
25%	         : 1사분위수 (데이터의 하위 25%)
50% (median) : 중앙값 (데이터의 중간값)
75%	         : 3사분위수 (데이터의 상위 25%)
max	         : 최댓값
'''
dfDeposit.describe() # df의 수치형(numeric) 데이터에 대한 요약 통계
```

```python
'''
- 총 행(row) 수 : df에 포함된 총 데이터 행 개수
- 각 열(column)의 이름
- 데이터 타입 : int64, float64, object, 등
- 각 열에 포함된 유효값(Non-Null 값) 개수 : 결측값이 아닌 데이터의 개수
- 데이터프레임의 메모리 사용량
'''
dfDeposit.info() # df의 전체적인 정보 요약
# => max_limit, dcls_end_day는 결측치가 많아서 활용하기 어려움
```

```python
dfDeposit # 데이터프레임의 기본 정보 (기본적으로 20개 출력)
```

```python
bar = dfDeposit.groupby(['kor_co_nm']).size().reset_index(name='count')
# df에서 `kor_co_nm`` 열을 기준으로 데이터 그룹화
# -> 해당 그룹에 포함된 행의 개수 계산
# -> 시리즈(series) 반환
# -> df로 변환하고, 새로 생성된 열의 이름을 `count`로 지정
bar # 출력
# => 중복 값 개수 확인 가능 & 그래프에 활용
```

```python
import seaborn as sns # 데이터 시각화 위한 Python 라이브러리
import matplotlib.pyplot as plt # 그래프나 차트를 그릴 때 사용하는 라이브러리
plt.rc("font", family = "Malgun Gothic") # Matplotlib의 기본 글꼴을 Windows 시스템의 맑은 고딕(Malgun Gothic)으로 설정 -> 한국어 폰트 깨지지 않도록
# 그래프 꾸미기
plt.title("은행 별 상품 개수", fontsize=16)
plt.xlabel("데이터 개수(count)", fontsize=12)
plt.ylabel("은행명(kor_co_nm)", fontsize=12)

sns.set(font="Malgun Gothic", rc={"axes.unicode_minus":False}, style='white') #  Seaborn 그래프에서 사용할 기본 폰트를 맑은 고딕으로 설정 & 음수 기호가 깨지는 문제를 방지 & 그래프의 배경 흰색
sns.barplot (data=bar, y='kor_co_nm', x='count', hue='kor_co_nm', palette='mako') # 수평 막대그래프를 생성 & 그래프에 사용할 데이터프레임은 bar & kor_co_nm 열의 값을 y축에 표시 & count 열의 값을 x축에 표시
```

```python
dfDeposit.groupby(['kor_co_nm','fin_prdt_nm', 'join_way'])['mtrt_int'].count().reset_index()
# df를 kor_co_nm(은행명), fin_prdt_nm(상품명), join_way(가입 방법)으로 그룹화 -> 정렬되어 그룹화 객체로 반환
# -> 값은 각 그룹에서 mtrt_int 열의 값 개수 계산 -> 결측값(Null)이 아닌 행의 수
# -> 인덱스를 초기화하고, 계산된 값을 데이터프레임 형태로 반환
```
